```python
# Enlaces

# [Dataset PIB (México, Italia, Ecuador)](https://datos.bancomundial.org/indicador/NY.GDP.MKTP.CD)
# [Dataset Tasa de Desempleo (México, Italia, Ecuador)](https://datos.bancomundial.org/indicador/SL.UEM.TOTL.NE.ZS)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

df_gdp = pd.read_csv(r'C:\Users\User\OneDrive\Escritorio\API_NY.GDP.MKTP.CD_DS2_es_csv_v2_3432.csv', skiprows=4)
df_unemployment = pd.read_csv(r'C:\Users\User\OneDrive\Escritorio\API_SL.UEM.TOTL.NE.ZS_DS2_es_csv_v2_8744.csv', skiprows=4)

paises = ['México', 'Italia', 'Ecuador']
anios = [str(anio) for anio in range(2000, 2024)]

# Filtrar datos de PIB y Tasa de Desempleo
datos_pib = df_gdp[df_gdp['Country Name'].isin(paises)][['Country Name'] + anios]
datos_desempleo = df_unemployment[df_unemployment['Country Name'].isin(paises)][['Country Name'] + anios]

# Cambiar nombres de columnas para facilitar el acceso
datos_pib.set_index('Country Name', inplace=True)
datos_desempleo.set_index('Country Name', inplace=True)
datos_pib = datos_pib.T
datos_desempleo = datos_desempleo.T

# Convertir los índices a tipo 'str' para que sean consistentes al graficar
datos_pib.index = datos_pib.index.astype(str)
datos_desempleo.index = datos_desempleo.index.astype(str)

# Función para imprimir tabla con los datos de PIB y Tasa de Desempleo
def imprimir_tabla(pais):
    # Unir los datos de PIB y Tasa de Desempleo en un solo DataFrame
    tabla = pd.DataFrame({
        'Año': datos_pib.index,
        'PIB (USD)': datos_pib[pais].values,
        'Tasa de Desempleo (%)': datos_desempleo[pais].values
    })
    # Mostrar la tabla
    print(f"\nTabla de PIB y Tasa de Desempleo para {pais}:")
    print(tabla)

#  Visualización de los datos del PIB y Tasa de Desempleo
for pais in paises:
    # Imprimir la tabla de PIB y Tasa de Desempleo
    imprimir_tabla(pais)

    # Gráfico del PIB
    plt.figure(figsize=(12, 6))
    plt.plot(datos_pib.index, datos_pib[pais], label='PIB (USD)', color='blue')
    plt.title(f'PIB en {pais} a lo largo del tiempo')
    plt.xlabel('Año')
    plt.ylabel('PIB (USD)')
    plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor visibilidad
    plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
    plt.grid(True)
    plt.legend()
    plt.show()

    # Gráfico de la Tasa de Desempleo
    plt.figure(figsize=(12, 6))
    plt.plot(datos_desempleo.index, datos_desempleo[pais], label='Tasa de Desempleo (%)', color='red')
    plt.title(f'Tasa de Desempleo en {pais} a lo largo del tiempo')
    plt.xlabel('Año')
    plt.ylabel('Tasa de Desempleo (%)')
    plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor visibilidad
    plt.grid(True)
    plt.legend()
    plt.show()

    # Calcular y mostrar la correlación entre PIB y Tasa de Desempleo a lo largo de los años
    pib_series = pd.to_numeric(datos_pib[pais], errors='coerce')
    desempleo_series = pd.to_numeric(datos_desempleo[pais], errors='coerce')
    correlacion = pib_series.corr(desempleo_series)
    print(f"Correlación entre PIB y Tasa de Desempleo en {pais} a lo largo del tiempo: {correlacion:.2f}")
```

```python
#Como científico de datos, al analizar la relación entre el Producto Interno Bruto (PIB) y la tasa de desempleo en México, Italia y Ecuador, podemos concluir que, aunque todos los países muestran una correlación (Algunos mas ligera que otros) entre ambas variables, las dinámicas varían notablemente. 
#En México, la débil correlación positiva sugiere que el crecimiento económico casi no está asociado y no está generando una disminución significativa en el desempleo, lo que podría estar relacionado con factores como la informalidad laboral o la insuficiencia de políticas públicas de empleo. 
#Por otro lado, Italia y Ecuador presentan correlaciones negativas, lo que indica una relación más directa entre el aumento del PIB y la reducción del desempleo. Ecuador refleja una correlación más fuerte, seguramente debido a su estructura económica que responde de manera
#más eficiente al crecimiento económico en términos de creación de empleo. 

# Estas diferencias apuntan a que, aunque el crecimiento económico es fundamental, como se administre y funcione cada país va a determinar la relación entre ambas variables.
```
